```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

colnames = ['year', 'name',"class","age","county"]#just fake names
data = pd.read_csv("data_banknote_authentication.csv",names=colnames) 

item1 = pd.Series(data['year']) 
item2 = pd.Series(data['name']) 
item3 = pd.Series(data['class']) 
item4 = pd.Series(data['age']) 
likes = pd.Series(data['county']) 

x = np.array([item1,item2,item3,item4])
y = np.array([likes])

print("arr - ",x[0:10])

print("arr - ",y)

print("inpust data \n\n",x)

wmat1 = np.random.rand(30,4)                   
b1 = np.random.rand(30,1)      

wmat2 = np.random.rand(1,30)                   
b2 = np.random.rand(1,1)      

print("weigths -",wmat1 )


def sigmoid(x):
    return 1/(1+np.exp(-x))
    
    
def relu(x):
    
    return np.maximum(x, 0)    

error = []
for i in range(10000):
    
    #forwoard propagation
    z1 = np.dot(wmat1,x)+b1             
    a1 = relu(z1)
    
    z2 = np.dot(wmat2,a1)+b2
    a2 = sigmoid(z2)
    
    
    
    #back propergation
    dz2 = a2-y 
    dw2 = np.dot(dz2,a1.T)/1372    
    db2 = np.sum(dz2,axis  = 1,keepdims=True)/1372
    
    dz1 = np.dot(wmat2.T,dz2)*a1
    dw1 = np.dot(dz1,x.T)/1372
    db1 = np.sum(dz1,axis=1,keepdims=True)/1372
    
    wmat1 = wmat1 - 0.001*dw1
    b1 = b1 - 0.001*db1
    
    wmat2 = wmat2 - 0.001*dw2
    b2 = b2 -0.001*db2
    
    err = np.sum(dz2,axis =1)
    error.append(err)
    
    print("\nerr*******-",i,"***************************-",err)

plt.plot(range(10000), error, label='error')

plt.xlabel('itarations')
plt.ylabel('error')

plt.title("error vs how many times trained")

plt.legend()

plt.show()

plt.scatter(range(1372), a2[0], label='yhat')


plt.xlabel('bank note fake or not')
plt.ylabel('index')

plt.title("final yhat")

plt.legend()

plt.show()
```
