```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline


data = load_iris()

x = data.data[0:100]
y = data.target[0:100]


wmat1 = np.random.rand(5,4)                   
b1 = np.random.rand(5,1)      

wmat2 = np.random.rand(1,5)                   
b2 = np.random.rand(1,1)      


def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return np.maximum(x, 0)

error = []
for i in range(2000):
    
    #forwoard propagation
    z1 = np.dot(wmat1,x.T)+b1             
    a1 = relu(z1)
    
    z2 = np.dot(wmat2,a1)+b2
    a2 = sigmoid(z2)
    
    
    
    #back propergation
    dz2 = a2-y 
    dw2 = np.dot(dz2,a1.T)/100     
    db2 = np.sum(dz2,axis  = 1,keepdims=True)/100
    
    dz1 = np.dot(wmat2.T,dz2)*a1
    dw1 = np.dot(dz1,x)/100
    db1 = np.sum(dz1,axis=1,keepdims=True)/100
    
    wmat1 = wmat1 - 0.001*dw1
    b1 = b1 - 0.001*db1
    
    wmat2 = wmat2 - 0.001*dw2
    b2 = b2 -0.001*db2
    
    
    err = np.sum(dz2,axis =1)
    error.append(err)
    
    print("\nerror*****************************************-",err)
    
    
plt.plot(range(2000), error, label='error')

plt.xlabel('itarations')
plt.ylabel('error')

plt.title("error vs how many times trained")

plt.legend()

plt.show()

plt.scatter(range(100), a2[0], label='yhat')

plt.xlabel('flower type ')
plt.ylabel('index')

plt.title("final yhat")

plt.legend()

plt.show()
    
print("a2 - \n",a2)
    
```
